```python
import numpy as np 
import pandas as pd
```

```python
panorama_parsed = pd.DataFrame({"title":np.load('../input/panorama-parsing/parsed_full.npy')})
panorama_parsed['is_fake'] = 1
panorama_parsed
```

```python
lenta_parsed = pd.DataFrame({"title":np.load('../input/lenta-news-kontyp/lenta_titles.npy', allow_pickle = True)})
lenta_parsed['is_fake'] = 0
lenta_parsed
```

```python
train = pd.read_csv("../input/fake-news-task/train.tsv", sep = '\t')
train
```

```python
train = panorama_parsed.append(lenta_parsed, ignore_index = True) \
                .append(train, ignore_index = True) \
                .sample(frac = 1, random_state = 1) \
                .reset_index(drop = True)
```

```python
train, validate, test = np.split(train, [int(.9*len(train)), int(.95*len(train))])
```

```python
train.reset_index(drop = True).to_feather('train.feather')
validate.reset_index(drop = True).to_feather('validate.feather')
test.reset_index(drop = True).to_feather('test.feather')
```
